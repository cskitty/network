```python
# larger figure size
plt.figure(3,figsize=(16,16)) 
nx.draw(G, with_labels = True)
```

# TV Series

## TV Series: Game of Thrones


Time-slice based networks, where all interactions are agglomerated every 10 scenes (denoted 'ts10').
* The nodes are represented by the names of the corresponding characters and the distance between two character names is inversely proportional to the weight of the corresponding edge: the closer they are at some point of the story, the more they interact then.

* The color intensity of each character name is proportional to the local strength of the corresponding character at the moment considered.


```python
def readsceneGoT(s):
    return nx.read_graphml('Data/GoT_dyn_ts10/' + s)
    
files = os.listdir("Data/GoT_dyn_ts10")

graphs = []
for f in files:
    graphs.append(readsceneGoT(f))

# larger figure size
plt.figure(3,figsize=(16,16)) 
nx.draw(graphs[2], with_labels = True)
```

# Movies

## Movie: Star Wars

```python
import json
 
def readjsonstarwars(n):
    G = nx.Graph()
    f = open('Data/starwars/starwars-episode-' + str(n) + '-interactions.json')
    data = json.load(f)
 
    c = 0
    for i in data['nodes']:
        G.add_node(c, name=i["name"])
        c += 1
    for i in data['links']:
        G.add_edge(i["source"], i["target"], weight=i["value"])
        c += 1
    f.close()
    describe(G)
    return G
```

```python
sw = readjsonstarwars(1)
plt.figure(3,figsize=(16,16)) 
nx.draw(sw, with_labels = True)
```

```python
sw = readjsonstarwars(2)
plt.figure(3,figsize=(16,16)) 
nx.draw(sw, with_labels = True)
```
