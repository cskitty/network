## Networks for George R. R. Martin's "A Song of Ice and Fire" Saga
Source: https://github.com/mathbeveridge/asoiaf


```python
import csv
import networkx as nx
import matplotlib.pyplot as plt

def describe(G):
    print("# nodes = ", len(G.nodes))
    #for node in G.nodes(data=True):
        #print(node)
    print("# edges = ", len(G.edges()))
    #for edge in G.edges(data=True):
        #print(edge)

G = nx.Graph()

def readnodes(b):
    csv_file = open("Data/asoiaf/data/asoiaf-book" + str(b) + "-nodes.csv")
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        line_count += 1
        if line_count == 1: 
            continue
        G.add_node(row[0])
        

def readedges(b):
    csv_file = open("Data/asoiaf/data/asoiaf-book" + str(b) + "-edges.csv")
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        line_count += 1
        if line_count == 1: 
            continue
        G.add_edge(row[0], row[1], weight=row[3])
    
for i in range(1, 6):
    readnodes(i)
    readedges(i)
    describe(G)
    
describe(G)
```

```python
# larger figure size
plt.figure(3,figsize=(16,16)) 
nx.draw(G, with_labels = True)
print("# nodes = ", len(G.nodes))
print("edges = ", G.edges() )
```

# TV Series - Game of Thrones : S01 - S02 - S03 (E01--02).


Time-slice based networks, where all interactions are agglomerated every 10 scenes (denoted 'ts10').
* The nodes are represented by the names of the corresponding characters and the distance between two character names is inversely proportional to the weight of the corresponding edge: the closer they are at some point of the story, the more they interact then.

* The color intensity of each character name is proportional to the local strength of the corresponding character at the moment considered.


```python
import networkx as nx
import csv
import pandas as pd
import matplotlib.pyplot as plt

def describe(G):
    print("# nodes = ", len(G.nodes))
    for node in G.nodes(data=True):
        print(node)
    print("# edges = ", len(G.edges()))
    for edge in G.edges(data=True):
        print(edge)
        
    #plt.figure(3,figsize=(16,16)) 
    #nx.draw(G, with_labels = True)


G1 = nx.read_graphml('GoT_dyn_ts10/GoT_S01E01_001.graphml')
G2 = nx.read_graphml('GoT_dyn_ts10/GoT_S01E01_002.graphml')
Gn = nx.read_graphml('GoT_dyn_ts10/GoT_S02E08_388.graphml')
```
